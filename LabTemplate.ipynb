# **Function**

```
def f(x,y):
  return x + y 

print('f(3,4) =',f(3,4))

def printMatrix(A):
   rows = A.shape[0]
   cols = A.shape[1]
   for i in range(0,rows):
      for j in range(0,cols):
         print(A[i,j]," ", end='')
      print()

printMatrix(A)
printMatrix(A.transpose())
```

# Triangular **mesh**

```
# Generate Delaunay mesh
Nx = 5
Ny = 5
hx = 1.0/(Nx-1)
hy = 1.0/(Ny-1)

x = np.ones((Nx*Ny))
y = np.ones((Nx*Ny))

for i in range(Nx):
  for j in range(Ny):
    x[i*Nx + j] = j*hx
    y[i*Nx + j] = i*hy

mesh = tri.Triangulation( x, y )
print('vertices = ', mesh.x, mesh.y)
print('edges =', mesh.edges)
print('neighbor triangles =', mesh.neighbors)

plt.figure(2)
plt.figure(figsize=(10,10))
plt.triplot(mesh, 'bo-', lw=1)
plt.title('Delaunay mesh')
```
